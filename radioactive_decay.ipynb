## The radioactive isotope $^{208}\text{Tl}$ (thallium 208) decays to stable $^{208}\text{Pb}$ (lead 208) with a half-life of 3.053 minutes. Suppose we start with a sample of 1000 thallium atoms. Simulate the decay of thallium atoms over time, mimicking the randomness of that decay.

- ### Initialize the number of thallium atoms to 1000 and the number of lead atoms to 0.
- ### Divide time into time steps of $\Delta t = 1$ s.
- ### During each time step, loop through all the thallium atoms and, for each one, determine whether or not it decays according to the probability $p(\Delta t) = 1 - 2^{-\Delta t/\tau}$.
- ### Work out the total number of thallium atoms that have decayed during the time step and update the counts of remaining thallium and lead atoms accordingly.

```python
N_Tl = 1000   # Number of Tl atoms
N_Pb = 0      # Number of Pb atoms

tmax = 1000  # Maximum time (in seconds)
dt = 1      # Step size

tau = 3.053*60   # Half-life of Tl in seconds

p = 1 - 2**(-dt/tau)

t_values = np.arange(0, tmax+dt, dt)


N_Tl_values = []
N_Pb_values = []

# Progress through time in steps of 1 seconds
for t in t_values:
  N_Tl_values.append(N_Tl)
  N_Pb_values.append(N_Pb)

  # Loop through currently avaialble Tl atoms and check if each atom decays or not
  num_Tl_decays = 0

  for i in range(N_Tl):
    if np.random.rand() < p:
      num_Tl_decays += 1

  N_Tl -= num_Tl_decays
  N_Pb += num_Tl_decays
```

```python
plt.plot(t_values, N_Tl_values, label='Tl')
plt.plot(t_values, N_Pb_values, label='Pb')
plt.xlabel('time (s)')
plt.ylabel('Number of atoms')
plt.legend()
```
